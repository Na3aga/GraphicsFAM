## Creating dictionary with function values and putting into file

```python
import numpy
x_start = -10
x_stop = 20
inp = "x*x*x + 4"


def output_function(x1, x2, equation):
  result = []
  for x in numpy.arange(x1,x2,0.1):
    y = eval(equation,{'x':x})
    result.append((round(x,2),round(y,2)))
  print(result)
  return result

#runnng

table_values = output_function(x_start,x_stop,inp)
with open('tablefunction.txt', 'w') as f:
    for point in table_values:
        f.write('%s %s\n' % point)
```

## Reading from the file 

```python
table_values = []

# open file and read the content in a list
with open('tablefunction.txt', 'r') as f:
    for line in f:
        # remove linebreak which is the last character of the string
        pnt = line[:-1]
        # add item to the list
        table_values.append(tuple(pnt.split(' ')))

print(table_values)
```

## Import Plotly 


```python
import plotly.graph_objects as go
```

```python
x_values = [point[0] for point in table_values]
y_values = [point[1] for point in table_values]


fig = go.Figure(
    data=[go.Scatter(x = x_values, y = y_values, line=dict(color="crimson"))],
    layout=dict(title=dict(text="A Chart"))
)
fig.show()
```
